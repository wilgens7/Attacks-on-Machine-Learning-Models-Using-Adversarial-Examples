#Global Variables

* This is for convenience so that we can easily map the results of the classifications to their proper label.

```
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
```

#Util Functions

* All the util functions to help prevent code repetition. All are self explanatory. 

```

def show_image(image, label=None):
  
  class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

  plt.figure(figsize=(15,15))
  plt.subplot(5,5,1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(image, cmap=plt.cm.binary)
    # The CIFAR labels happen to be arrays, 
    # which is why you need the extra index
  if label != None:
   plt.xlabel(class_names[label[0]])
  plt.show()


def get_label(model_prediction):
  """given the prediction of a model this function 
  prints out the associated label"""

  return class_names[np.argmax(model_prediction)]

def get_confidence(classifier, example):

  """Gets confidence levels for predictions"""
  
  pred = classifier.predict(example)
  label = np.argmax(pred,axis=1)
  confidence = np.max(pred, axis=1)

  print('label:',class_names[label[0]],'confidence:',confidence[0])

def get_attack_name(attack):
  """Returns attack name"""
  return str(attack).split('.')[3]

def get_attack_type(attack):
  """Returns attack type"""
  return str(attack).split('.')[2]
  
def get_defense_name(defense):
  """Returns defense name"""
  return str(defense).split('.')[3]

def get_defense_type(defense):
  """Returns defense type"""
  return str(defense).split('.')[2]
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers

substitute_model = keras.models.Sequential()
#substitute_model.add(layers.Dense(3072))
substitute_model.add(layers.Conv2D(32, (3, 3), activation='relu', padding='same'))
substitute_model.add(layers.MaxPooling2D((2, 2)))
substitute_model.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
substitute_model.add(layers.MaxPooling2D((2, 2)))
substitute_model.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
substitute_model.add(layers.Flatten())
substitute_model.add(layers.Dense(64, activation='relu'))
substitute_model.add(layers.Dense(10))
substitute_model.add(layers.Softmax())
```

```
substitute_model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

substitute_model.build(input_shape=(None,32,32,3))
```

###Jacobian


```
##Create Jacobean function

def jacobian_for(model, x):
  """
  Calculate jacobian of logits wrt input.
  """
  x = tf.convert_to_tensor(x,dtype=tf.float32)
  x = tf.expand_dims(x,0)

  with tf.GradientTape(watch_accessed_variables=False, persistent=True) as tape:
   
    tape.watch(substitute_model.get_layer(index=0).variables)

    y = model(x,training=False)

  return tf.squeeze(tape.jacobian(y, 
                                  x, 
                                  parallel_iterations=100, 
                                  experimental_use_pfor=False))
  
```
